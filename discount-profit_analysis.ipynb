## Discount and Profit Analysis

### Analysis Overview

This analysis investigates the relationship between **Discount** and **Profit** using real transactional data from a retail dataset. The goal is to determine whether offering higher discounts leads to higher or lower profitability.

### Supporting Theory

This analysis is grounded in the principle of **cost-volume-profit (CVP) analysis**, which is commonly used in managerial accounting. According to this theory, a company's profitability depends on the interplay between the price at which goods are sold, the cost to produce or acquire those goods, and the volume of sales. Discounts lower the selling price, which can reduce the contribution margin per unit. If not compensated by a significant increase in sales volume, this will lead to lower profits or even losses ([Kotler & Keller, 2016](https://www.pearson.com)).

### Methods and Libraries Used

* **Libraries:**

  * `pandas` for data loading and manipulation ([pandas.pydata.org](https://pandas.pydata.org))
  * `matplotlib.pyplot` and `seaborn` for data visualization ([matplotlib.org](https://matplotlib.org), [seaborn.pydata.org](https://seaborn.pydata.org))
  * `scipy.stats.linregress` for performing linear regression analysis ([docs.scipy.org](https://docs.scipy.org))

* **Methods:**

  * Scatter plot: Visualize the distribution and pattern between discount and profit
  * Linear regression: Fit a trendline and compute the correlation between the two variables


```python
!pip install scipy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Load data
df = pd.read_csv("cleaned_superstore.csv")
```

```python

# scatterplot discount-profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Discount", y="Profit", alpha=0.6)
plt.title("Relationship between Discount and Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True)

# linear regresion
slope, intercept, r_value, p_value, std_err = linregress(df["Discount"], df["Profit"])
x_vals = df["Discount"]
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color="red", label=f"y = {slope:.2f}x + {intercept:.2f}")
plt.legend()

plt.tight_layout()
plt.show()

# correlation p-value
print(f"Korelasi (r): {r_value:.4f}")
print(f"P-value: {p_value:.4f}")
```

## Analysis of the Relationship between Discount and Profit


### What Does the Chart Show?

This chart illustrates the relationship between **discounts given to customers** and the **profit** earned from sales. Each dot represents a single sales transaction.

* **Horizontal axis (X)** shows the **discount** as a percentage (e.g., 0.2 = 20%).
* **Vertical axis (Y)** shows the **profit** made after applying the discount.
* **Blue dots** represent the actual transaction data.
* **Red line** is the trend line (linear regression) showing the direction of the relationship between discount and profit.

### What Does It Mean?

1. **The higher the discount, the lower the profit tends to be.**

   * We can see that as discounts increase, many profit values drop below zero (indicating losses).
   * The red downward-sloping line confirms this pattern, showing that, on average, discounts reduce profit.

2. **Trendline equation: y = -249.05x + 67.56**

   * This means: for every 1-unit increase in discount (e.g., from 10% to 20%), the average profit drops by about 249 units of currency.
   * The negative sign indicates a **negative** correlation.

3. **Why can large discounts cause losses?**

   * Because excessive discounts can cut into the selling price margin.
   * If the selling price is reduced too much, it may not cover the cost of goods sold, leading to a loss.

### Conclusion

* Discounts may help attract buyers, but they need to be applied carefully.
* Excessive discounts tend to **reduce profit** and may even lead to **financial loss**.
* Therefore, discount strategies should strike a balance between appealing to customers and preserving profit margins.
